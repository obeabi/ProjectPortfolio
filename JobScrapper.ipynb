## Step 1

```python
# install libraries
import pandas as pd
import numpy as np
import time
from bs4 import BeautifulSoup
from selenium import webdriver
print("library installation complete!")
```

```python
# Define Selenium options
driver = webdriver.Chrome()

options = webdriver.ChromeOptions()
options.add_argument('--incognito')
options.add_argument('--headless')

driver.get('https://ca.indeed.com/')
time.sleep(5)

print("Selenium options established!")
```

## Step 2

```python
# Establish and input serach parameters for job scraping through the following actions

button_one = driver.find_element_by_xpath('//*[@id="text-input-what"]').send_keys('analytics')
time.sleep(5)



button_two = driver.find_element_by_xpath('//*[@id="text-input-where"]').send_keys(u'\ue009' + u'\ue003')

button_three = driver.find_element_by_xpath('//*[@id="text-input-where"]').send_keys(u'\ue009' + u'\ue003')

button_four = driver.find_element_by_xpath('//*[@id="text-input-where"]').send_keys(u'\ue009' + u'\ue003')

time.sleep(5)

button_five = driver.find_element_by_xpath('//*[@id="text-input-where"]').send_keys('british columbia')

time.sleep(5)

button_six = driver.find_element_by_xpath('//*[@id="whatWhereFormId"]/div[3]/button').click()

time.sleep(5)
```

## Step 3

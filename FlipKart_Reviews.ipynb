## Step 1

```python
# Install libraries
import os
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import matplotlib.pyplot as plt

%matplotlib inline

print("Libraries Installed Successfully!")
```

```python
# Provide url to be scraped and erach string
partial_url         = "https://www.flipkart.com/search?q="
searchString = 'samsung'
url  = partial_url + searchString 
print("The url to be scraped is"+ ' '+ url)
```

## Step 2

```python
# send request to url and parse web content
uClient = uReq(url)
Page = uClient.read()  # data dump
uClient.close()
data_html = bs(Page, "html.parser")

print(data_html)
```

```python
# Obtain box that captures each product
bigboxes = data_html.findAll("div", {"class": "bhgxx2 col-12-12"})
del bigboxes[0:3]  # ensure the last value corresponds to the exact class name in html document
box = bigboxes[0]
print(box)

print('======================================================')
print("The length of the variable box is", len(box))
```

```python
# Get link for individual product
urrl = "https://www.flipkart.com"
productLink = urrl + box.div.div.div.a['href']
print(productLink)
```

```python
# Access and Parse product link 
prodRes = requests.get(productLink)
prodRes.encoding = 'utf-8'
prod_html = bs(prodRes.text, "html.parser")

# scrape each box for reviews
commentboxes = prod_html.find_all('div', {'class': "_3nrCtb"})
```

```python
# Create csv file for saving results
filename = searchString + ".csv"
fw = open(filename, "w")
headers = "Product, Customer Name, Rating, Heading, Comment \n"
fw.write(headers)
```
